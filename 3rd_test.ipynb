```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("merged_data_3rd_test.csv")
data
```

```python
# Plotting all bearing X and Y data against index
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
for column in data.columns:
        plt.plot(data.index, data[column], label=column, linestyle='-')
        # plt.plot(data.index, data[column], label=column, linestyle='--')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Bearing Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a list of colors for the plots
colors = ['b', 'orange', 'g', 'r']

# Create 8 different plots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))  # Create a 4x2 grid of subplots
fig.tight_layout(pad=4.0)  # Adjust the spacing between subplots

# Plot each pair of X and Y bearing data on separate subplots
for i, (ax, column) in enumerate(zip(axs.flatten(), data.columns)):
        ax.plot(data.index, data[column], label=column, color=colors[i], linestyle='-')
        ax.set_title(column)
        ax.set_xlabel('Index')
        ax.set_ylabel('Value')
        ax.grid(True)

# Add legend to the last subplot
axs.flatten()[-1].legend()

plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("merged_data_3rd_test.csv")

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Apply PCA
pca = PCA()
pca.fit(scaled_data)

# Get the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Print explained variance ratio
print("Explained Variance Ratio:")
print(explained_variance_ratio)

# Plot explained variance ratio
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
# plt.title('Scree Plot')
plt.show()
```

```python

# Load the dataset
data = pd.read_csv("merged_data_3rd_test.csv")

# Convert the data to a numpy array
X = data.to_numpy()

# Perform Singular Value Decomposition (SVD)
U, S, VT = np.linalg.svd(X, full_matrices=False)

# Print the shapes of U, S, and VT matrices
print("Shape of U:", U.shape)
print("Shape of S:", S.shape)
print("Shape of VT:", VT.shape)
```
